```python
#SETUP
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

if os.environ.get("GROQ_API_KEY"):
    print("Bro API KEY Variable exists")
else:
    raise ValueError("GROQ_API_KEY not found")

from langchain_core.messages import HumanMessage, SystemMessage, AIMessage

llm_groq = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
```

# **GUIDING IN PROMPTS**

```python
from langchain_core.prompts import PromptTemplate

result = llm_groq.invoke(
    """
    Tell me a joke.
    
    Generate the output in key-value pair format
    with the following keys:
    setup, punchline
    """
)

result.content
```

# **PYDANTIC MODEL**

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    setup: str = Field(description="The setup for the joke")
    punchline: str = Field(description="The punchline for the joke")
```

```python
llm_structured_output = llm_groq.with_structured_output(
    llm_schema
)

result = llm_structured_output.invoke(
    "Tell me a joke"
)

result.punchline
```

# **TYPEDDICT**

```python
from typing import TypedDict

class llm_schema_td(TypedDict):
    setup: str
    punchline: str
```

```python
llm_structured_typed_dict = (
    llm_groq.with_structured_output(llm_schema_td)
)

result = llm_structured_typed_dict.invoke(
    "Tell me a joke"
)

result
```
