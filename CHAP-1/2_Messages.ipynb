# **Messages**

```python
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv

# Load variables from .env
load_dotenv()

if os.environ.get("GROQ_API_KEY"):
    print("Bro API KEY Variable exists")
else:
    raise ValueError("GROQ_API_KEY not found")
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_groq import ChatGroq

llm_groq = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)

my_messages = [
    SystemMessage(
        content="You are a gen-z assistant who always answers in a fun and human-like way"
    ),
    HumanMessage(content="Bro, tell me a fun fact")
]

llm_groq.invoke(my_messages).content
```

# **PROMPTS**

```python
from langchain_core.prompts import PromptTemplate

user_input = input("Enter a topic for fun fact : ")

dynamic_prompt = PromptTemplate.from_template(
    "Write a fun fact about {topic}"
)

ready_prompt = dynamic_prompt.invoke({
    "topic": user_input
})

llm_groq.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {tone} assistant who replies naturally like a human"),
    ("user", "Write a fun fact about {topic}.")
])

user_input = input("Enter a topic: ")
user_tone = input("Enter a tone: ")

ready_prompt = prompt_template.invoke({
    "topic": user_input,
    "tone": user_tone
})

llm_groq.invoke(ready_prompt.messages)
```
