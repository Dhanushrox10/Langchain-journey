# **SETUP**

```python
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

if os.environ.get("GROQ_API_KEY"):
    print("Bro API KEY Variable exists")
else:
    raise ValueError("GROQ_API_KEY not found")

from langchain_core.messages import (
    HumanMessage,
    SystemMessage,
    AIMessage
)

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_groq = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
```

# **FIRST CHAIN**

```python
# TASK - 1 [Prompt]

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])
```

```python
# TASK - 2 [LLM]

llm_groq = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
```

```python
# TASK - 3 [Str Parser]

str_parser = StrOutputParser()
```

# **BY MANUAL WAY**

```python
template = prompt_template.invoke({
    "input": "What is the capital of France?"
})

res = llm_groq.invoke(template)

final_result = res.content

print(final_result)
```

# **BY CHAIN WAY**

```python
chain = prompt_template | llm_groq | str_parser

chain.invoke({
    "input": "What is the capital of France?"
})
```

# **RunnableSequence Version**

```python
from langchain_core.runnables import RunnableSequence

chain_1 = RunnableSequence(
    prompt_template,
    llm_groq,
    str_parser
)

chain_1.invoke({
    "input": "What is the capital of France?"
})
```
