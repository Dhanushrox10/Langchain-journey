# **SETUP**

```python
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

if os.environ.get("GROQ_API_KEY"):
    print("Bro API KEY Variable exists")
else:
    raise ValueError("GROQ_API_KEY not found")

from langchain_core.messages import (
    HumanMessage,
    SystemMessage,
    AIMessage
)

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_groq = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
```

# **CHAIN WITH CUSTOM RUNNABLE**

```python
# TASK - 1 [Prompt]

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])
```

```python
# TASK - 2 [LLM]

llm_groq = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
```

```python
# TASK - 3 [Str Parser]

str_parser = StrOutputParser()
```

# **CUSTOM RUNNABLE**

```python
# TASK - 4 [Custom Runnable]

from langchain_core.runnables import RunnableLambda

def dictionary_maker(text: str) -> dict:

    return {"text": text}

dictionary_maker_runnable = RunnableLambda(
    dictionary_maker
)
```

```python
# TASK - 5 [Template For Post]

prompt_post = ChatPromptTemplate.from_messages([
    ("system", "You're a social media post generator."),
    (
        "human",
        "Create a LinkedIn post for the following text: {text}"
    ),
])
```

```python
# TASK - 6 [Generate the Post]

llm_groq = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)
```

```python
# TASK - 7 [Str Parser]

str_parser = StrOutputParser()
```

```python
# CHAIN

chain = (
    prompt_template
    | llm_groq
    | str_parser
    | dictionary_maker_runnable
    | prompt_post
    | llm_groq
    | str_parser
)

chain.invoke("What is the capital of France?")
```
